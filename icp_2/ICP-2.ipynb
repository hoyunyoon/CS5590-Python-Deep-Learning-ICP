```python
# Hoyun Yoon ICP-2

def problem_1():
    cont = 'YES'
    inch_heights = []
    cm_heights = []

    # Loop until the customer is done entering heights
    while cont[0].lower() == 'y':

        # Ask user to enter the height in inches and add it to the inches list
        height = float(input("Enter height in inches: "))
        inch_heights.append(height)

        # Convert the height to cm and add it to the cm list
        cm = float(height * 2.54)
        cm_heights.append(cm)

        # Ask user if they have more data to add
        cont = input("Do you want to add more data (Y / N): ")

        # Keep asking for valid input
        while (cont[0].lower() != 'y') and (cont[0].lower() != 'n'):
            cont = input("Do you want to add more data (Y / N): ")

    # Print out the heights in inches and centimeters
    print("\nHeights in inches")
    print(inch_heights)
    print("Heights in cm")
    print(cm_heights)


def full_name(first, last):
    full = first + " " + last
    return full


def string_alternative(name):
    alt = ""

    # Loop through name and concatenate every other letter
    for i in range(len(name)):
        if i % 2 == 0:
            alt += name[i]

    return alt


def problem_2():
    # Ask user for their first and last names
    first = input("Enter first name: ")
    last = input("Enter last name: ")

    # Send the strings to the full name function
    full = full_name(first, last)

    # Send full name to string alternative function
    alt = string_alternative(full)

    # Display the full name and the alternative string
    print("Your full name is ", full, ".")
    print("Alternative string for your first name is ", alt, ".")


def problem_3():
    # Open the input text file and create an output text file
    file = open("input.txt", 'r')
    output_file = open("output.txt", 'w')
    line = file.readline()
    word_count = {}

    while line != "":
        words = line.split()
        for i in words:
            if i not in word_count:
                word_count[i] = 1
            else:
                word_count[i] = word_count[i] + 1
        line = file.readline()

    # Write the word count information to the output file
    output_file.write("Word Count:\n")
    for i in word_count.keys():
        output_file.write(i + ": " + str(word_count[i]) + "\n")

    # Close the files
    file.close()
    output_file.close()


if __name__ == '__main__':
    problem_1()
    problem_2()
    problem_3()
```
